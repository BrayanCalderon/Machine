```python
TD = np.array([[2,3,0,3,7],
               [0,5,5,0,3],
               [5,0,7,3,3],
               [3,1,0,9,9],
               [0,0,7,1,3],
               [6,9,4,6,0]])
L = np.array([5,2,3,6,4,3]).T
print(TD)
print(L)
```

```python
#TD Matriz que corresponde a la cantidad de veces que una palabra T aparece en un documento D, por ejemplo en la posicion 0,0
#la palabra en la posicion 0 aparece 2 veces en el documento 0
```

```python
total_words = np.sum(TD)
PTD = TD/total_words
print(PTD)
```
